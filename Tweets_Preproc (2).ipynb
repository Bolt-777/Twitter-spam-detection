```python
proc_data = []

def cleanup_tw(data):
    split_data = re.sub(',', '\n', data, count=12)
    return split_data.splitlines()

with open("C:\\Users\\hp pc\\Downloads\\Twitter-1KS-10KN-data\\Twitter-1KS-10KN-data\\NormalData\\New_Tweet.csv","r") as fp:
    lines = fp.read().split('\n')
    
proc_data = [cleanup_tw(data) for data in lines]
proc_data[:2]
```

```python
proc_df = pd.DataFrame(proc_data, columns=['tweet_id', 'user_id', 'tweet creation time', 'favourited', 'in_reply_to_status_id', 'in_reply_to_user_id', 'in_reply_to_screen_name', 'latitude', 'longitude', 'source', 'retweet', 'truncated', 'text'])
proc_df.head()
```

```python
href_list = []
rel_list = []

for index in range(len(proc_df['source'])):
    #print(proc_df['source'][index])

    try:
        href = re.search('(href=)(["a-zA-Z:/.]+)', proc_df['source'][index]).group(2).strip('"')
        #print(href)
        href_list.append(href)
        
        rel = re.search('(rel=)(["a-zA-Z]+)', proc_df['source'][index]).group(2).strip('"')
        #print(rel)
        rel_list.append(rel)
        
    except AttributeError:
        #print("Web")
        href_list.append('Web')
        rel_list.append('Web')
    except:
        print(f"Exception on data {index}: ", proc_df['source'][index])
        href_list.append('None')
        rel_list.append('None')        
```

```python
proc_df['href'] = href_list
proc_df['rel'] = rel_list

proc_df.head()
```
